### 2.42 Distribution Plots (distplot)

For numerical data we should always understand the distribution. Histogram and KDE are two related plots for visualizing distributions.

```python
# tips is the dataframe; total_bill is the variable
sns.distplot(tips.total_bill)
```

Distplot has three primary settings, histogram, kde, and rug. These can be turned on and off. Default has kde and histogram on. We can also specify color.

```python
sns.distplot(tips.total_bill, hist=True, kde=False, rug=False, color='g')
```

```python
# we will get an error unless we dropna
sns.distplot(titanic.age.dropna(), kde=True,rug=True)
```

### 2.43 Distributions (boxplot, violinplot)

Boxplot was invented by John Tukey. It shows the interquartile range and also outliers. Outliers have to be understood and treated with care as part of any analysis.

```python
sns.boxplot(tips.total_bill)
```

```python
sns.boxplot(titanic.age)
```

```python
# violin plot shows boxplot and the distribution
plt.figure(figsize=(10,10))
sns.violinplot(tips.total_bill, color="g")
sns.violinplot(tips.tip)
```

```python
# comparing distributions when there is a categorical variable
sns.boxplot(x="tip",y="sex",data=tips)
```

### 2.44 Bivariate Data (jointplot, pairplot)

```python
# middle is scatter plot
sns.jointplot(x="total_bill",y="tip" , data=tips)
```

```python
# hexagon more points
sns.jointplot(x="total_bill",y="tip" , data=tips, kind="hex")
```

```python
# regression more points
sns.jointplot(x="total_bill",y="tip" , data=tips, kind="reg")
```

```python
# regression more points
sns.jointplot(x="total_bill",y="tip" , data=tips, kind="kde")
```

```python
# does jointplot but for every possible combinations of numerical value
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue="sex")
```

```python
sns.jointplot(x="age",y="fare",data=titanic)
```

### 2.45 Categorical Data (Counts)

```python
sns.countplot(x="sex",data=tips)
```

```python
# aggregate typically the mean
sns.barplot(x="sex",y="total_bill",data=tips)
```
